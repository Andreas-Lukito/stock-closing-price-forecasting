# Stock Closing Price Forecasting

## Import Libraries

```python
# Common Python Libraries
import numpy as np
import pandas as pd
import os
import joblib
from tqdm import tqdm

# Data
import yfinance as yf

# Project Path
project_path = "../"
```

## Pull Data

```python
snp_500 = yf.Ticker("^GSPC")
```

```python
data = snp_500.history(interval="1d", start="2010-11-30", end="2025-11-30")
```

```python
data_path = os.path.join(project_path, "data/raw_data.csv")
data.to_csv(data_path)
```
